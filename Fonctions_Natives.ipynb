### `all(iterable)`
Renvoie `True` si tous les éléments de iterable sont vrais (ou s'il est vide). Équivaut à :


```
def all(iterable):
    for element in iterable:
        if not element:
            return False
    return True
```



````python
depenses = [23, 566, 78]
comparaisons = [6544, " ", -34, 677, True, 5>14]
````

````python
# Vérifier si tous les éléments de ces itérables sont vraies
print(all(depenses))
print(all(comparaisons))
````

### `any(iterable)`
Renvoie True si au moins un élément de iterable est vrai. False est renvoyé dans le cas où iterable est vide. Équivaut à :


```
def any(iterable):
    for element in iterable:
        if element:
            return True
    return False
```



````python
revenus = [7885, 566, 78]
comparaisons = [6544, " ", -34, 677, True, 5>14]
inegalites = [False, False, False, False]
````

````python
# Vérifier si au moins un des éléments de ces itérables est vrai
print(any(revenus))
print(any(comparaisons))
print(any(inegalites))
````

````python
N = int(input())
integers_str = input().split()

all_positive = all([int(number)>=0 for number in integers_str])
any_palindromic = any([number == number[::-1] for number in integers_str] )

if 0 < N < 100:
    if all_positive and any_palindromic:
        print(True)
    else:
        print(False)
````

### `bool(x=False)`
Renvoie une valeur booléenne, c'est-à-dire True ou False. x est converti à l'aide de la procédure standard de vérification de la vérité.

#### Vérification standard de la vérité
Tout objet peut être comparé à une valeur booléenne, généralement dans une condition if ou while ou en tant qu'opérande des opérations booléennes ci-dessous.

Par défaut, un objet est considéré comme vrai à moins que sa classe ne définisse une méthode __bool__() qui renvoie False ou une méthode __len__() qui renvoie zéro, lorsqu'elle est appelée avec l'objet. 1 Voici la plupart des objets intégrés considérés comme faux :

constantes définies comme fausses : None et False

zéro de tout type numérique : 0, 0.0, 0j, Decimal(0), Fraction(0, 1)

les chaînes vides et les collections : '', (), [], {}, set(), range(0)

Les opérations et fonctions natives dont le résultat est booléen renvoient toujours 0 ou False pour false et 1 ou True pour true, sauf indication contraire (exception importante : les opérations booléennes or et renvoient toujours l'un de leurs opérandes).

### `bytearray()`
class `bytearray(source=b'', encoding, errors)`
Renvoie un nouveau tableau d'octets. La classe bytearray est une séquence mutable de nombres entiers dans l'intervalle 0 ≤ x < 256. Il possède la plupart des méthodes des séquences variables, décrites dans Types de séquences mutables, ainsi que la plupart des méthodes de la classe `bytes`, voir Opérations sur les `bytes` et `bytearray`.

Le paramètre optionnel source peut être utilisé pour initialiser le tableau de plusieurs façons :

* si c'est une chaîne, vous devez aussi donner le paramètre encoding pour l'encodage (et éventuellement errors). La fonction bytearray() convertit ensuite la chaîne en octets via la méthode str.encode() ;

* si c'est un entier, le tableau a cette taille et est initialisé d'octets null ;

* si c'est un objet conforme à l'interface tampon, un tampon en lecture seule de l'objet est utilisé pour initialiser le tableau ;

* si c'est un itérable, il doit itérer sur des nombres entier dans l'intervalle 0 <= x < 256, qui sont utilisés pour initialiser le contenu du tableau.

* Sans argument, un tableau vide est créé.



````python
compliment = "Python is beautiful and easy to learn"
iban = [123, 45, 79]
````

````python
print(bytearray(compliment, encoding='utf-16'))
print(bytearray(iban))
````

### `bytes()`

En conséquence, les arguments du constructeur sont les mêmes que pour bytearray().

*   Les objets bytes (octets) peuvent aussi être créés à partir de littéraux, voir Littéraux de chaînes de caractères et de suites d'octets.

*   Les objets octets contiennent des données lisibles par une machine et nous pouvons stocker un objet octet directement dans le stockage secondaire.

````python
compliment = "Python is a high-level language"
vitesse = 12
````

````python
print(bytes(compliment, encoding='utf-16'))
print(bytes(vitesse))
````

La méthode __next__() de l'itérateur donné par enumerate() renvoie un n-uplet contenant un compte (démarrant à start, 0 par défaut) et les valeurs obtenues de l'itération sur iterable.

https://pythonbasics.org/enumerate/

Enumérer une liste

````python
fruits = ['pomme', 'orange', 'mangue', 'prune']
````

````python

for index in range(len(fruits)):
    print(index, fruits[index])
````

### `eval(expression, globals=None, locals=None)`

Cette fonction permet l'exécution dynamique de code Python. expression doit être soit une chaîne soit un objet code. Si c'est une chaîne, elle est d'abord analysée en une suite d'instructions Python qui sont ensuite exécutés (sauf erreur de syntaxe). Si c'est un objet code, il est simplement exécuté. Dans tous les cas, le code fourni doit être valide selon les mêmes critères que s'il était un script dans un fichier. Gardez en tête que les mots clés `nonlocal`, `yield` et `return` ne peuvent pas être utilisés en dehors d'une fonction, même dans du code passé à eval(). La fonction renvoie None.

Les arguments sont : une chaîne, et optionnellement des locales et des globales. S'il est fourni, `globals` doit être un dictionnaire. S'il est fourni, `locals` peut être n'importe quel objet mapping.

L'argument expression est analysé et évalué comme une expression Python (techniquement, une condition list) en utilisant les dictionnaires `globals` et `locals` comme espaces de nommage global et local.

Cette fonction peut aussi être utilisée pour exécuter n'importe quel objet code (tels que ceux créés par `compile()`). Dans ce cas, donnez un objet code plutôt qu'une chaîne. Si l'objet code a été compilé avec l'argument mode à `'exec'`, `eval()` renvoie None.

````python
eval('45 - 6 +2')
````

````python
salaire_mensuel = 6300
eval('salaire_mensuel + 300')
````

````python
# vous ne pouvez pas faire des assignations de valeur avec eval
salaire_mensuel = 6300
eval('salaire_annuel = salaire_mensuel + 300')
````

### `filter(function, iterable)`
Construit un itérateur depuis les éléments d'iterable pour lesquels function renvoie `True`. iterable peut aussi bien être une séquence, un conteneur qui prend en charge l'itération, ou un itérateur. Si function est None, la fonction identité est prise, c'est-à-dire que tous les éléments faux d'iterable sont supprimés.

Notez que `filter(function, iterable)` est l'équivalent du générateur `(item for item in iterable if function(item))` si function n'est pas None, et de `(item for item in iterable if item)` si function est None.

Voir `itertools.filterfalse()` pour la fonction complémentaire qui donne les éléments d'iterable pour lesquels function renvoie `False`.

````python
# Filtrer les nombres pairs


def verifie_parite(nombre):
    if nombre % 2 == 0:
        return True
    return False


# Application
elements = [244, 56, 78, 22, 21, 7855, 980]

elements_filtres = list(filter(verifie_parite, elements))

print(elements_filtres)
````

````python
# Filtrer les villes qui commencent par 'M'

def check_start_with_M(name):
    if name.startswith('M'):
        return True
    return False

villes = ['Lisbonne', 'Madrid', 'Macau', 'Stockolm', 'Magdebourg', 'Berlin']

villes_en_M = list( filter(check_start_with_M, villes))
villes_en_M
````

````python
# Filter avec une fonction anonyme (Lambda)
# identifier les nombres pairs

nombres = [244, 56, 78, 22, 21, 7855, 980]
nombres_pairs = list( filter(lambda x: (x%2 == 0), nombres) )

print(nombres_pairs)
````

````python
# Filter avec  None comme fonction
# identifier les éléments qui renvoient `False` si on leur applique `bool()`

corbeille = [False, [], "Bonjour", 34, True, 34.7, "", -21, 0, 15, (4, 6)]
corbeille_nettoyee = list( filter(None, corbeille) )

print(corbeille_nettoyee)
````

### `float(x=0.0)`
Renvoie un nombre a virgule flottante depuis un nombre ou une chaîne `x`.

Si l'argument est une chaîne, elle doit contenir un nombre décimal, éventuellement précédé d'un signe, et pouvant être entouré d'espaces. Le signe optionnel peut être `'+'` ou `'-`'. Un signe `'+'` n'a pas d'effet sur la valeur produite. L'argument peut aussi être une chaîne représentant un `NaN` (Not-a-Number ou pas un nombre en français), l'infini positif, ou l'infini négatif. Plus précisément, l'argument doit se conformer à floatvalue tel que défini la grammaire suivante, après que les espaces en début et fin de chaîne aient été retirées :

* sign        ::=  "+" | "-"
* infinity    ::=  "Infinity" | "inf"
* nan         ::=  "nan"
* digitpart   ::=  `!digit` (["_"] `!digit`)*
* number      ::=  [digitpart] "." digitpart | digitpart ["."]
* exponent    ::=  ("e" | "E") ["+" | "-"] digitpart
* floatnumber ::=  number [exponent]
* floatvalue  ::=  [sign] (floatnumber | infinity | nan)

Pour un objet Python général `x`, `float(x)` est délégué à `x.__float__()`. Si `__float__()` n'est pas défini alors il est délégué à `__index__()`.

````python
# Code
print(float('+1.23'))
print(float('   -12345\n'))
print(float('1e-003'))
print(float('+1E6'))
print(float('-Infinity'))
print(float('1234_564_56.90'))
````

### `format(value, format_spec='')`
Convertit une valeur en sa représentation « formatée », contrôlée par format_spec. L'interprétation de format_spec dépend du type de la valeur. Cependant, il existe une syntaxe standard utilisée par la plupart des types natifs : Mini-langage de spécification de format.

Par défaut, format_spec est une chaîne vide. Dans ce cas, appeler cette fonction a généralement le même effet qu'appeler str(value).


````python
# Exemples
presentation1 = "Mon nom est {fvorname} {fname}, J'ai {age} ans".format(fvorname="John", fname = "Doe", age = 44)
presentation2 = "Mon nom est {0} {1}, J'ai {2} ans".format("Clara", "Doe", 22)
presentation3 = "Mon nom est {1} {0}, J'ai {2} ans".format("Max", "John", 16)
presentation4 = "Mon nom est {} {}, J'ai {} ans".format("Brio", "Doe", 12)

identite = ("Bonnie", "Doe", 42)
presentation5 = "Mon nom est {} {}, J'ai {} ans".format(*identite)
````

````python
print(presentation1)
print(presentation2)
print(presentation3)
print(presentation4)
print(presentation5)
````

````python
# Autres exemples

# affiche avec 2 chiffres après la virgule
description = "Cet ordinateur coûte {prix:.2f} Euros"
print( description.format(prix = 123455.4567))

# affiche avec un séparateur de milliers '_'
description = "Cet ordinateur coûte {prix:_.2f} Euros"
print( description.format(prix = 123455.4567))


# aligne à droite avec espaces à gauche pour un emplacement de 15 caractères
description = "Cet ordinateur coûte {prix:>15_.2f} Euros"
print( description.format(prix = 123455.4567))

# aligne au centre avec * padding à gauche et à droite, et ',' en séparateur de milliers
description = "Cet ordinateur coûte {prix:*^15,.4f} Euros" # center
print( description.format(prix = 123455.4567))

# aligne à gauche avec zéro padding à droite
description = "Cet ordinateur coûte {prix:0<15,.2f} Euros"
print( description.format(prix = 123455.4567))
````
