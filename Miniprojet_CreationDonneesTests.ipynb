### Importer les modules nécessaires

```python
import random
import pandas as pd
import datetime
```

### Données manuelles

```python
types_de_tests = [
                'Functional',
                'Reliability',
                'Performance',
                'Operability',
                'Security',
                'Compatibility',
                'Maintenanbility',
                'Transferability'
        ]

types_de_resultats = [
                'Passed',
                'Failed',
                'Skipped',
                'Error'
        ]

dates_possibles = [
                datetime.date(2023, 12, 10),
                datetime.date(2023, 12, 11),
                datetime.date(2023, 12, 12),
                datetime.date(2023, 12, 13)
        ]
```

## Paramètres d'entrée du programme

```python
nombre_de_tests = 10
type_de_fichier_produit = 'csv'
```

## Générer les ids

```python
ids = []
for id in range(nombre_de_tests): # [0, 1, 2, 3, 4, ..., 45, 46, 47, 48, 49]
    ids.append(id)

```

```python
# ids = list(range(nombre_de_tests))
```

```python
print(ids)
```

### Générer des noms de tests

```python
prefix = "Test_"
noms_des_tests = []
for id in ids:
    nom_test = prefix + str(id)
    noms_des_tests.append(nom_test)
```

```python
"Tests_ " + str(12)
```

```python
print(noms_des_tests)
```

### Générer des types de tests


```python
categories_de_tests = random.choices(types_de_tests, weights=[50, 10, 20, 5, 4, 2, 2, 2], k=nombre_de_tests)
```

```python
print(categories_de_tests)
```

## Générer des dates d'exécution

```python
dates_execution = random.choices(dates_possibles, k=nombre_de_tests)
```

```python
dates_execution
```

### Générer les temps d'exécution

```python
# solution 1
temps_executions = random.choices([3, 100, 600, 700], k=nombre_de_tests)
temps_executions
```

```python
# solution avec distribution Gaussienne
temps_executions = []
for temps in range(nombre_de_tests):
    temp = random.gauss(mu=500.0, sigma=200.0) + 300 # random.randint(1, 600)
    temps_executions.append(temp)

temps_executions
```

### Générer les résultats de test


```python
help(random.choices)
```

```python
types_de_resultats
```

```python
resultats_de_tests = random.choices(types_de_resultats,
                                   weights=[90, 5, 3, 2 ],
                                   k=nombre_de_tests)

resultats_de_tests
```

### Dictionnaires de donnees

```python
donnees_tests = {
    "id": ids,
    "Test_name": noms_des_tests,
    'Test_type': categories_de_tests,
    'Exec_date': dates_execution,
    'Exec_duration': temps_executions,
    'Test_result': resultats_de_tests
}
```

```python
from pprint import pprint
pprint(donnees_tests)
```

## Solution avec open


```python
from pprint import pprint
pprint(donnees_tests)
```

```python
donnees_globales = []

en_tete = ["id", "Test_name",
           "Test_type", "Exec_date",
           "Exec_duration", "Test_result"]

donnees_globales.append(en_tete)

for i in range(nombre_de_tests):
    ligne = [
        str(donnees_tests["id"][i]),
        str(donnees_tests["Test_name"][i]),
        str(donnees_tests["Test_type"][i]),
        str(donnees_tests["Exec_date"][i]),
        str(donnees_tests["Exec_duration"][i]),
        str(donnees_tests["Test_result"][i])
    ]

    donnees_globales.append(ligne)

donnees_globales
```

```python
# Ecrire dans un fichier
with open('results_v6.csv', mode='w') as fichier_csv:
    for ids, ligne in enumerate(donnees_globales):
        fichier_csv.write( ",".join(ligne) + "\n")
```

```python
# Ecrire dans un fichier
with open('results_v7.csv', mode='w') as fichier_csv:
    for ligne in donnees_globales:
        ligne = ",".join( map(str, ligne)) + "\n"
        fichier_csv.write(ligne)
        #print(idx, ligne)
```
